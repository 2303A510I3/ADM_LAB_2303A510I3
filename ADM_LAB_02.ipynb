```python
import pandas as pd
```

```python
data = pd.read_csv('/content/titanic.csv')
summary = data.describe()
print(summary)
print(data.describe(include=['object']))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['Age'].dropna(), bins=10,edgecolor='black', color='skyblue')
plt.title('Titanic')
plt.xlabel('PassengersId')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
data['Pclass'].value_counts().plot(kind='bar',color='lightcoral')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data['Age'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data[['PassengerId','Survived','Pclass','Age']])
plt.show()
```

```python
correlation_matrix = data[['PassengerId','Survived','Pclass','Age']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.countplot(x='Age', data=data, palette = 'Set2')
plt.title('Age disturbution by class')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='Age', y='Fare', data=data)
plt.title('Pclass vs Fare')
plt.xlabel('Pclass')
plt.ylabel('Fare')
```

```python
sns.violinplot(x='Pclass', y='Fare', data=data, palette = 'muted')
plt.title('Age disturbution by passenger class')
plt.show()
```
